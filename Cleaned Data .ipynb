# Cleaning dataset with python

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"E:\pandas\Machine Learning\Projects\Car_details(U).csv")
data
```

```python
# 1. Identify the maximum and minimum values of the selling_price column.
#(Calculate the average of these two values and use this average tofill all missing values in selling_price.)
```

```python
max_price = data["selling_price"].max()
min_price = data["selling_price"].min()
```

```python
avg_value = (max_price+min_price)/2        # Average Calculation
avg_value
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#filling missing values
data["selling_price"].fillna(avg_value,inplace = True)
```

```python
#2. Handle the km_driven and owner_count columns using an appropriate real-world data-cleaning technique.
#Justify the method used for each column based on data distributionand practical usage.
```

```python
# filling KM missing value with median
```

```python
data["km_driven"].fillna(data["km_driven"].median(),inplace= True)
```

```python
# filling owner missing value with mode
```

```python
data["owner_count"].fillna(data["owner_count"].mean(),inplace= True)
```

```python
#3. For the km_category column, 50% of the missing values should be filled with the most frequently occurring category.
#The remaining 50% of missing values should be filled with the value "High".
 
```

```python
data.head(2)
```

```python
# Finding Mode
```

```python
mode_km = data["km_category"].mode()[0]
mode_km
```

```python
# Missing Index
```

```python
missing_index = data[data["km_category"].isna()].index
missing_index
```

```python
half = len(missing_index)//2
half
```

```python
# first half
```

```python
data.loc[missing_index[:half],"km_category"] = mode_km
```

```python
# secong half
```

```python
data.loc[missing_index[half:],"km_category"] = "High"
```

```python
#4. Fill missing values in the price_category column using the most frequently occurring category in the dataset.
#If it is not possible to determine whether the category is generated manually or automatically, replace missing values with "Unknown".
```

```python
if data["price_category"].mode().empty:
    data["price_category"].fillna("Unknown",inplace=True)
else:
    data["price_category"].fillna(data["price_category"].mode()[0],inplace=True)
```

```python
#5. For the fuel column, distribute all missing values equally among the existing unique fuel types.
#Each unique fuel value must be assigned the same number of missing records.
```

```python
fuel_type = data["fuel"].dropna().unique()
fuel_type
```

```python
missing_index = data[data["fuel"].isna()].index
missing_index
```

```python
n_types = len(fuel_type)
n_types
```

```python
for i , idx in enumerate(missing_index):
    data.loc[idx,"fuel"] = fuel_type[i%n_types]
```

```python
# filling Null values in Transmission Column (Extra Activity)
```

```python
data["transmission"].fillna(data["transmission"].mode()[0],inplace=True)
```
